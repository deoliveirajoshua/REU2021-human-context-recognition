```python
import pandas as pd
import matplotlib.pyplot as plt
from pylab import *
%matplotlib inline
```

```python
#dynamic_data = pd.read_csv('../model_outputs/DynamicGan.csv')
#constant_data = pd.read_csv('../model_outputs/Test Gan.csv')

dynamic_data = pd.read_csv('../model_outputs/1k_5_2_dynamic_GAN.csv')
constant_data = pd.read_csv('../model_outputs/1k_5_2_constant_GAN.csv')

median_dynamic = dynamic_data[['Median Wasserstein']]
median_constant = constant_data[['Median Wasserstein']]
mean_dynamic = dynamic_data[['Mean Wasserstein']]
mean_constant = constant_data[['Mean Wasserstein']]
```

```python
median_diff = median_dynamic - median_constant
mean_diff = mean_dynamic - mean_constant

epochs = list(range(1, 952))
```

```python
plt.figure(figsize = (15, 5), dpi = 300)
subplot(1, 2, 1)
plt.plot(epochs, median_diff, "b-", color = '#2994d6')
plt.hlines(0, 1, 1000, colors = ['#000000'], linewidths = 1)
plt.xlabel('Epochs')
plt.title('Median Wasserstein Distance (Dynamic - Conventional)')
subplot(1, 2, 2)
plt.plot(epochs, mean_diff, "b-", color = '#2994d6')
plt.hlines(0, 1, 1000, colors = ['#000000'],  linewidths = 1)
plt.xlabel('Epochs')
plt.title('Mean Wasserstein Distance (Dynamic - Conventional)')
plt.show()

 
    
```
