```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
all_users = pd.read_csv('../model_outputs/All User F-1 Scores.csv')
one_user = pd.read_csv('../model_outputs/Single User F-1 Scores.csv')

all_mean = np.mean(all_users.iloc[:,:-1].values, axis = 0)
one_mean = np.mean(one_user.iloc[:,:-1].values, axis = 0)

runs = [k for k in range(100)] + [k for k in range(100)]
label = ['All Users'] * 100 + ['One User'] * 100

all_f1 = np.concatenate((all_mean, one_mean))
```

```python
f1 = pd.DataFrame({'Run Number': runs, 'All User F1': all_f1, 'Label': label})
f1
```

```python
sns.scatterplot(data = f1, x = 'Run Number', y = 'All User F1', hue = 'Label')
plt.xlabel('Random Trial Number')
plt.ylabel('F1 Score')
plt.ylim(0,1)
plt.legend(loc = 'center right', title = 'Legend')
plt.title('Tailored vs. Untailored HAR for 6 Activities')
sns.despine()
plt.savefig('tailored.png', dpi = 1080, transparent = True)
```
